```python
'''
Part 1:
- filter the entire dataset, include only patients with one condition
- limit the max number of occurences for any given patient to 100
'''

# Part 1, filter the entire dataset, include only patients with one condition

cnv = df[(df['class'] == 'CNV') & (df['class'] != 'NORMAL') & (df['class'] != 'DME') & (df['class'] != 'DRUSEN')]
dme = df[(df['class'] == 'DME') & (df['class'] != 'NORMAL') & (df['class'] != 'CNV') & (df['class'] != 'DRUSEN')]
drusen = df[(df['class'] == 'DRUSEN') & (df['class'] != 'NORMAL') & (df['class'] != 'CNV') & (df['class'] != 'DME')]

# limit the max number of occurences for any given patient to 100

cnv = cnv[~cnv['patient_id'].isin(cnv['patient_id'].value_counts()[cnv['patient_id'].value_counts() > 20].index)]
dme = dme[~dme['patient_id'].isin(dme['patient_id'].value_counts()[dme['patient_id'].value_counts() > 20].index)]
drusen = drusen[~drusen['patient_id'].isin(drusen['patient_id'].value_counts()[drusen['patient_id'].value_counts() > 20].index)]
```

```python
'''
Part 2:
- from the filtered lists, randomly sample unique patients, and select ALL their images to include
- Ensure that the sum of images from each condition is 1000
'''


'''
Part 3:

- split into train (80%), val (10%), test (10%)
- ensure no patient appears in more than one set
- ensure each set has a balanced representation of each condition

- train: 800 CNV, 800 DME, 800 DRUSEN
- val: 100 CNV, 100 DME, 100 DRUSEN
- test: 100 CNV, 100 DME, 100 DRUSEN
'''


def split_data(cnv, dme, drusen, length):
    # access global df's, and modify as needed
    n_cnv, n_dme, n_drusen = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

    while (len(n_cnv) + len(n_dme) + len(n_drusen)) < length:
        # cnv
        if len(n_cnv) < (length // 3):
            unique_patients = cnv['patient_id'].unique()
            selected_patient = random.choice(unique_patients)
            patient_images = cnv[cnv['patient_id'] == selected_patient]
            n_cnv = pd.concat([n_cnv, patient_images])
            cnv = cnv[cnv['patient_id'] != selected_patient]  # Remove selected patient from pool

        # dme
        if len(n_dme) < (length // 3):
            unique_patients = dme['patient_id'].unique()
            selected_patient = random.choice(unique_patients)
            patient_images = dme[dme['patient_id'] == selected_patient]
            n_dme = pd.concat([n_dme, patient_images])
            dme = dme[dme['patient_id'] != selected_patient]  # Remove selected patient from pool

        # drusen
        if len(n_drusen) < (length // 3):
            unique_patients = drusen['patient_id'].unique()
            selected_patient = random.choice(unique_patients)
            patient_images = drusen[drusen['patient_id'] == selected_patient]
            n_drusen = pd.concat([n_drusen, patient_images])
            drusen = drusen[drusen['patient_id'] != selected_patient]  # Remove selected patient from pool
    
    return (n_cnv, n_dme, n_drusen), (cnv, dme, drusen)

# split into train (80%), val (10%), test (10%)

(t_cnv, t_dme, t_drusen), (cnv, dme, drusen) = split_data(cnv, dme, drusen, 2400)
(v_cnv, v_dme, v_drusen), (cnv, dme, drusen) = split_data(cnv, dme, drusen, 300)
(te_cnv, te_dme, te_drusen), (cnv, dme, drusen) = split_data(cnv, dme, drusen, 300)





# f_cnv, f_dme, f_drusen = pd.DataFrame(), pd.DataFrame(), pd.DataFrame()

# while (len(f_cnv) + len(f_dme) + len(f_drusen)) < 3000:
#     # cnv
#     if len(f_cnv) < 1000:
#         unique_patients = cnv['patient_id'].unique()
#         selected_patient = random.choice(unique_patients)
#         patient_images = cnv[cnv['patient_id'] == selected_patient]
#         f_cnv = pd.concat([f_cnv, patient_images])
#         cnv = cnv[cnv['patient_id'] != selected_patient]  # Remove selected patient from pool

#     # dme
#     if len(f_dme) < 1000:
#         unique_patients = dme['patient_id'].unique()
#         selected_patient = random.choice(unique_patients)
#         patient_images = dme[dme['patient_id'] == selected_patient]
#         f_dme = pd.concat([f_dme, patient_images])
#         dme = dme[dme['patient_id'] != selected_patient]  # Remove selected patient from pool

#     # drusen
#     if len(f_drusen) < 1000:
#         unique_patients = drusen['patient_id'].unique()
#         selected_patient = random.choice(unique_patients)
#         patient_images = drusen[drusen['patient_id'] == selected_patient]
#         f_drusen = pd.concat([f_drusen, patient_images])
#         drusen = drusen[drusen['patient_id'] != selected_patient]  # Remove selected patient from pool

# print(f"CNV: {len(f_cnv)}, DME: {len(f_dme)}, DRUSEN: {len(f_drusen)}, TOTAL: {len(f_cnv) + len(f_dme) + len(f_drusen)}")
```

```python
# step 1 -> generate retrieval captions

'''
template 1:
1. depicted condition
2. depicted urgency
3. not non-depicted condition 1
4. not non-depicted condition 2
5. (hybrid) not non-depicted condition 1 OR 2 and depicted urgency

template 2:
1. depicted condition
2. depicted urgency
3. not non-depicted condition 1
4. not non-depicted condition 2
5. (..) depicted condition and depicted urgency
'''

def generate_retrieval_captions(disease_class, urgency, template):
    disease_map = {'CNV': 'Choroidal Neovascularization', 'DME': 'Diabetic Macular Edema', 'DRUSEN': 'Drusen'}
    urgency_map = {'YES': 'an urgent referral', 'NO': 'no urgent referral'}
    non_depicted = [disease_map[d] for d in disease_map.keys() if d != disease_class]
    first_condition = random.choice(non_depicted)
    second_condition = random.choice([d for d in non_depicted if d != first_condition])

    c1 = f"This OCT scan shows {disease_map[disease_class]}."
    c2 = f"This OCT scan requires {urgency_map[urgency]}."
    c3 = f"This OCT scan does not show {first_condition}."
    c4 = f"This OCT scan shows no evidence of {second_condition}."

    if template == '1':
        random_condition = random.choice(non_depicted)
        c5 = f"This OCT scan is without {random_condition} and requires {urgency_map[urgency]}."
    
    elif template == '2':
        c5 = f"This OCT scan exhibits {disease_map[disease_class]} and requires {urgency_map[urgency]}."

    captions = [c1, c2, c3, c4, c5]
    return captions

final_df['retrieval_captions'] = final_df.apply(lambda row: generate_retrieval_captions(row['disease class'], row['urgent referral'], random.choice(['1', '2'])), axis=1)

# s1 = subset.iloc[:1500].copy()
# s2 = subset.iloc[1500:].copy()

# s1['retrieval_captions'] = s1.apply(lambda row: generate_retrieval_captions(row['disease class'], row['urgent referral'], '1'), axis=1)
# s2['retrieval_captions'] = s2.apply(lambda row: generate_retrieval_captions(row['disease class'], row['urgent referral'], '2'), axis=1)

# oct_retrieval = pd.concat([s1, s2]).reset_index(drop=True)
```
